เนื่องจาก 0 คือผิดนัดชำระหนี้, 1 คือไม่ผิดนัดชำระหนี้ หากคำนวณค่าเฉลี่ยจะได้เป็นอัตราส่วนของผู้ขอกู้ที่ไม่ผิดนัดชำระหนี้

คำนวณอัตราส่วนของผู้ขอกู้ที่ไม่ผิดนัดชำระหนี้สำหรับแต่ละ grade:

```python
df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()
```

รวม DataFrames ของจำนวนแถว กับ DataFrame ของ อัตราส่วนของผู้ขอกู้ที่ไม่ผิดนัดชำระหนี้ให้เป็น DataFrame เดียว:

```python
df1 = pd.concat([df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].count(),
                df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()], axis = 1)
# Concatenate two DataFrames along the columns.
```

```python
df1
```

จากข้างบน ทั้ง 2 columns ที่ชื่อ grade มีชื่อเกรดที่ตรงกัน นั่นคือหลังจากทำการรวม DataFrames แล้วผลลัพธ์ยังคง match กัน

```python
df1 = df1.iloc[:, [0, 1, 3]]
df1
```

เปลี่ยนชื่อ columns ให้มีความชัดเจน คือ n_obs (number of observations) และ prop_good (proportion of good borrowers):

```python
df1.columns = [df1.columns.values[0], 'n_obs', 'prop_good']
df1
```

คำนวณอัตราส่วนผู้ขอกู้ของละเกรดต่อจำนวนผู้ขอกู้ทั้งหมด:

```python
df1['prop_n_obs'] = df1['n_obs'] / df1['n_obs'].sum()
df1
```

คำนวณจำนวนผู้ขอกู้ที่ผิดนัดชำระหนี้และไม่ผิดนัดชำระหนี้:

```python
df1['n_good'] = df1['prop_good'] * df1['n_obs']
df1['n_bad'] = (1 - df1['prop_good']) * df1['n_obs']
df1
```

คำนวณอัตราส่วนผู้ขอกู้ที่ผิดนัดชำระหนี้ของละเกรดต่อจำนวนผู้ขอกู้ทั้งหมดที่ผิดนัดชำระหนี้ และคำนวณอัตราส่วนผู้ขอกู้ที่ไม่ผิดนัดชำระหนี้ของละเกรดต่อจำนวนผู้ขอกู้ทั้งหมดที่ไม่ผิดนัดชำระหนี้:

```python
df1['prop_n_good'] = df1['n_good'] / df1['n_good'].sum()
df1['prop_n_bad'] = df1['n_bad'] / df1['n_bad'].sum()
df1
```

คำนวณ WoE (Weight of Evidence) สำหรับตัวแปร grade:

```python
df1['WoE'] = np.log(df1['prop_n_good'] / df1['prop_n_bad'])
df1
```

เรียงลำดับ Weight of Evidence จากต่ำไปสูง เพื่อให้กลุ่มผู้ขอกู้ที่มีอัตราการผิดนัดชำระหนี้สูงสุดอยู่ข้างบนสุด:

```python
df1 = df1.sort_values(['WoE'])
df1 = df1.reset_index(drop = True)
# Reset the index of a DataFrame and overwrite it.
df1
```

การคำนวณต่อไปนี้ไม่จำเป็นต้องคำนวณก็ได้ แต่สามารถคำนวณเพื่อเอาไว้ (เพื่อให้ทราบว่าจาก grade หนึ่งไปยังอีก grade หนึ่ง มีค่าเพิ่มขึ้นมากน้อยแค่ไหน)

สำหรับ columns ชื่อ prop_good และ WoE; จะทำการคำนวณผลต่างระหว่างค่าที่อยู่ในแถวนั้น ๆ กับค่าที่อยู่ในแถวก่อนหน้า; จากนั้นนำไปเก็บไว้ในตัวแปร diff_prop_good และ diff_WoE:

```python
df1['diff_prop_good'] = df1['prop_good'].diff().abs()
df1['diff_WoE'] = df1['WoE'].diff().abs()
df1
```

คำนวณ IV (Information Value):

$$\text{Information Value} = \sum_{i=1}^{k}\left[ \left( \%good-\%bad \right) \times \ln \left( \%good \over \%bad \right) \right]$$

```python
df1['IV'] = (df1['prop_n_good'] - df1['prop_n_bad']) * df1['WoE']
df1['IV'] = df1['IV'].sum()
df1
```

การสร้าง dummy variables:

ลักษณะกราฟแทบจะเป็น monotonic คือ ยิ่งอัตราดอกเบี้ยสูง ยิ่งมี Weight of Evidence ต่ำ

10 categories แรก กราฟมีลักษณะขึ้น ๆ ลง ๆ และมี Weight of Evidence ค่อนข้างสูงแต่จำนวนผู้ขอกู้ (n_obs) ไม่ได้มีมาก จึงยุบรวมได้เป็น `<=9.548`

จากนั้น Weight of Evidence ลดลงพอสมควร จึงยุบรวมได้เป็น `9.548-12.025`

จากนั้น Weight of Evidence ลดลงอีกพอสมควร จึงยุบรวมได้เป็น `12.025-15.74`

จากนั้น Weight of Evidence ลดลงอีกพอสมควร จึงยุบรวมได้เป็น `15.74-20.281`

หลังจาก 20.281 เป็นต้นไป Weight of Evidence ไม่ค่อยแตกต่างกันมาก และจำนวนผู้ขอกู้ (n_obs) ค่อยข้างน้อยด้วย จึงยุบรวมได้เป็น `>20.281`

ทำการสร้าง dummy variables ดังที่กล่าวไปได้ดังนี้:

```python
# '>20.281' (riskiest group) will be the reference category
df_inputs_prepr['int_rate:<=9.548'] = np.where((df_inputs_prepr['int_rate'] <= 9.548), 1, 0)
df_inputs_prepr['int_rate:9.548-12.025'] = np.where((df_inputs_prepr['int_rate'] > 9.548) & (df_inputs_prepr['int_rate'] <= 12.025), 1, 0)
df_inputs_prepr['int_rate:12.025-15.74'] = np.where((df_inputs_prepr['int_rate'] > 12.025) & (df_inputs_prepr['int_rate'] <= 15.74), 1, 0)
df_inputs_prepr['int_rate:15.74-20.281'] = np.where((df_inputs_prepr['int_rate'] > 15.74) & (df_inputs_prepr['int_rate'] <= 20.281), 1, 0)
df_inputs_prepr['int_rate:>20.281'] = np.where((df_inputs_prepr['int_rate'] > 20.281), 1, 0)
```
